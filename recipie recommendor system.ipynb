```python
def get_recommendation(recipe_id,recipes):
  ing = sim_ing_matrix[recipe_id].fillna(0).tolist()
  tag = sim_tag_matrix[recipe_id].fillna(0).tolist()

  final = []
  for i in range(len(ing)):
    score = 0.5*ing[i]+0.5*tag[i]
    final.append(score)

  df=pd.DataFrame()
  df['id']=recipes['Recipe Id']
  df['name']=recipes['TranslatedRecipeName']
  df['final']=final

  df = df.sort_values(by='final', ascending=False)

  return recipes['TranslatedRecipeName'][recipe_id],df['id'][1:6].tolist(),df['name'][1:6].tolist(),df['final'][1:6]
```
